```python
class Custom_Dataset(Dataset):
    def __init__(self, data):
        self.X = data.drop(['Alpha', 'Beta', 'Gamma', 'Delta', 'Class'], axis=1).values
        self.pre_y = data[['Alpha', 'Beta', 'Gamma', 'Delta']].values
        self.y = data['Class'].values
    
    def __getitem__(self, index):
        # return self.X[index, :], self.pre_y[index, :], self.y[index]
        return self.X[index, :],\
               {'alpha': self.pre_y[index, 0],
                'beta': self.pre_y[index, 1],
                'gamma': self.pre_y[index, 2],
                'delta': self.pre_y[index, 3]},\
               self.y[index]
    
    def __len__(self):
        return self.X.shape[0]

    def __getsize__(self):
        return (self.__len__())
```

#### model

```python
class NeuralNet(nn.Module):
    def __init__(self, n_neurons=[56, 64, 128, 64, 32, 16],
                 out_classes={'alpha': 4, 'beta': 3, 'gamma': 8, 'delta':4},
                 dropout_p=0.3, activation=nn.ReLU()):
        super(NeuralNet, self).__init__()
        
        net_layers = []
        # net_layers.append(nn.BatchNorm1d(n_neurons[0]))
        for i in range(1, len(n_neurons) - 1):
            net_layers.append(nn.BatchNorm1d(n_neurons[i-1]))
            net_layers.append(nn.Linear(n_neurons[i-1], n_neurons[i]))
            # net_layers.append(nn.BatchNorm1d(n_neurons[i]))
            net_layers.append(activation)
            net_layers.append(nn.Dropout(dropout_p))
        net_layers.append(nn.Linear(n_neurons[-2], n_neurons[-1]))
        
        self.fc = nn.Sequential(*net_layers)
        
        self.alpha = nn.Sequential(
            nn.BatchNorm1d(n_neurons[-1]),
            activation,
            nn.Linear(n_neurons[-1], out_features=out_classes['alpha'])
        )
        self.beta = nn.Sequential(
            nn.BatchNorm1d(n_neurons[-1]),
            activation,
            nn.Linear(n_neurons[-1], out_features=out_classes['beta'])
        )
        self.gamma = nn.Sequential(
            nn.BatchNorm1d(n_neurons[-1]),
            activation,
            nn.Linear(n_neurons[-1], out_features=out_classes['gamma'])
        )
        self.delta = nn.Sequential(
            nn.BatchNorm1d(n_neurons[-1]),
            activation,
            nn.Linear(n_neurons[-1], out_features=out_classes['delta'])
        )
        self.classifier = nn.Sequential(
            nn.BatchNorm1d(19),
            activation,
            nn.Linear(19, out_features=19),
            nn.BatchNorm1d(19),
            activation,
            nn.Linear(19, out_features=2)
        )
        
    def forward(self, x, multihead=True):
        x = self.fc(x)
        if (multihead==False):
            x = torch.concat([self.alpha(x), self.beta(x), self.gamma(x), self.delta(x)], dim=1)
            return self.classifier(x)
        return {'alpha': self.alpha(x),
                'beta': self.beta(x),
                'gamma': self.gamma(x),
                'delta': self.delta(x)
               }
```

```python
net = NeuralNet(activation=nn.LeakyReLU(), dropout_p=0.0)
net.to(device)
net.train()

def criterion(loss_func, preds, trues, binary_trues, multihead=True, device=device):
    # for normal criterion output
    if (multihead==False): return loss_func(preds.to(device), binary_trues.to(device))
    
    # for multilabel head criterion output
    losses = 0
    for i, key in enumerate(preds):
        losses += loss_func(preds[key].to(device), trues[key].to(device))
    return losses

for param in net.fc:
    if isinstance(param, nn.Linear):
        nn.init.kaiming_normal_(param.weight)
        nn.init.uniform_(param.bias)
```

```python
loss_func = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(net.parameters(), lr=0.0005)
train_list = []
eval_list = []
```

```python
def train(model, optimizer, loss_func,
          train_list, eval_list,
          train_dataset, eval_dataset,
          num_epochs=5, batch_size=8, print_times=4, verbose=True, multihead=True):
    
    print_every = len(train_dataset)//batch_size//print_times
    
    train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True, num_workers=4)
    eval_loader = DataLoader(eval_dataset, batch_size=batch_size, shuffle=True, num_workers=4)    
    
    for epoch in range(num_epochs):
        loss_train = 0
        counter = 0
        for X, y, z in tqdm.tqdm(train_loader):
            X, y, z = X.to(device), y, z.to(device)
            optimizer.zero_grad()
            preds = model.forward(X, multihead=multihead)
            
            loss_value = criterion(loss_func, preds, y, z, multihead=multihead)
            loss_train += loss_value.item()
            loss_value.backward()
            optimizer.step()
            
            counter += 1

            if (counter % print_every == 0):
                loss_eval = 0
                model.eval()
                with torch.no_grad():
                    for X, y, z in eval_loader:
                        X, y, z = X.to(device), y, z.to(device)

                        preds = model.forward(X, multihead=multihead)
                        loss_value = criterion(loss_func, preds, y, z, multihead=multihead)
                        loss_eval += loss_value.item()

                    if (verbose==True):
                        print(f'epoch {epoch} | ', f'{counter*batch_size}/{len(train_dataset)}', '\n',
                              f'train loss (mean per batch) - {loss_train/counter:.2f}', '\n',
                              f'eval loss (mean per batch) - {loss_eval/(len(eval_dataset)/batch_size):.2f}')

                    train_list.append(loss_train/counter)
                    eval_list.append(loss_eval/len(eval_dataset))
                model.train()
```

```python
for param in net.classifier.parameters(): param.requires_grad = False

train(net, optimizer, loss_func=loss_func,
      train_list=train_list, eval_list=eval_list,
      train_dataset=train_dataset, eval_dataset=eval_dataset,
      num_epochs=30, print_times=1, batch_size=8, multihead=True)

# for param in net.parameters(): param.requires_grad = False
# for param in net.classifier.parameters(): param.requires_grad = True

# train(net, optimizer, loss_func=loss_func,
#       train_list=train_list, eval_list=eval_list,
#       train_dataset=train_dataset, eval_dataset=eval_dataset,
#       num_epochs=10, print_times=1, batch_size=8, multihead=False)

for param in net.parameters(): param.requires_grad = True

train(net, optimizer, loss_func=loss_func,
      train_list=train_list, eval_list=eval_list,
      train_dataset=train_dataset, eval_dataset=eval_dataset,
      num_epochs=10, print_times=1, batch_size=8, multihead=False)
```

```python
multihead=False
outputs = []
eval_loader = DataLoader(eval_dataset, batch_size=16, shuffle=False, num_workers=4)

with torch.no_grad():
    for X, y, z in eval_loader:
        X, y, z = X.to(device), y, z.to(device)

        preds = F.softmax(net.forward(X, multihead=multihead), dim=1)[:, 1]
        outputs.append(preds.cpu().detach_().numpy())
        
outputs = np.concatenate(outputs)

from sklearn.metrics import log_loss, f1_score, accuracy_score, balanced_accuracy_score
print(log_loss(df_eval['Class'], outputs))
print(f1_score(df_eval['Class'], outputs>0.5))
print(accuracy_score(df_eval['Class'], outputs>0.5))
```
